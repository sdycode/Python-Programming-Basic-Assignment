1. Write a Python program to convert kilometers to miles?

```python
def convertKmtoMile(d) :
  return d*(1/1.60934)

dInMiles = convertKmtoMile(10)
print(dInMiles)
```

2. Write a Python program to convert Celsius to Fahrenheit?

```python
def Celsius_to_Fahrenheit(c) :
  return 32+ (9/5)*c

inFar = Celsius_to_Fahrenheit(100)
print(inFar)
```

4. Write a Python program to solve quadratic equation?

```python

import math 

def getRootsOFQuadaticEquations(a,b,c):
  def checkForRealORImaginary():
    if((b*b -4*a*c)>=0):
      return True,(b*b -4*a*c)
    else:
      return False, (b*b -4*a*c)
  print(checkForRealORImaginary())
  if(checkForRealORImaginary()[0]):
    print('Roots are real')
    root1 = (-b + math.sqrt(checkForRealORImaginary()[1]))/(2*a)
    root2 = (-b - math.sqrt(checkForRealORImaginary()[1]))/(2*a)
    return root1, root2
  else:
    print('Roots are Imaginary')
    return None, None


print(getRootsOFQuadaticEquations(5,8,2))
print(getRootsOFQuadaticEquations(5,3,2))

  
```

5. Write a Python program to swap two variables without temp variable?

```python
a = 10
b = 24
x = 20
y =12

def swap(p,q):
  if(p>q):

    p = (p-q)
    q = p+q
    p = q-p 
    
    return p, q
  else: 
    q = q-p
    p = p+q
    q = p-q 
    return p, q


print(swap(a,b))
swap(x, y)

```
