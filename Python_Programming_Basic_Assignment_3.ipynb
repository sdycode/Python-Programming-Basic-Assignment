1. Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python


def check(n):
  if(n<0):
    print('no is negative')
  elif(n>0):
    print('no is positive')
  else:
    print('no is Zero')

check(10)
check(0)
check(-45)
```

2. Write a Python Program to Check if a Number is Odd or Even?

```python
def checkEvenOdd(n):
  if(n & 1 == 1):
    print('no is Odd')
  else:
    print('no is Even')

checkEvenOdd(-5)
checkEvenOdd(12)

```

4. Write a Python Program to Check Prime Number?

```python

import math


def checkForPrime(n): 
  flag = False

  if n > 1:
    # check for factors
    for i in range(2, int(math.sqrt(n))):
        if (n % i) == 0:
            flag = True
            break
  else:
    flag = True

  
  if flag:
    print(n, "is not a prime number")
  else:
    print(n, "is a prime number")


checkForPrime(1)
checkForPrime(2)
checkForPrime(14)
checkForPrime(17)
checkForPrime(103)
checkForPrime(-53)
```
