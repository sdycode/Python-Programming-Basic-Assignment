1. Write a Python Program to Find the Factorial of a Number?

```python
def fact(n):
  if(n == 0):
    return 1;
  else:
    return n*fact(n-1)

print(fact(4))
print(fact(6))
print(fact(5))
```

2. Write a Python Program to Display the multiplication Table?

```python
def showMultiplicationTable(n): 
  for i in range(1,11):
    print(i*n)

for i in range(2,4):
  print(f'Table of {i}')
  showMultiplicationTable(i)
  
```

```python
def isArmstrong(n):
  sum = 0;
  m = n
  while(int(m/10>0)): 
    sum = sum + (m%10)**3
    m = int(m / 10)
    if(m/10 == 0) :
      sum = sum + (m%10)**3
  print(sum)
  if(sum == n):
    print("True")
    return True
  else:
    print("False")
    return False


isArmstrong(400)
isArmstrong(371)
isArmstrong(153)
```
