4. Write a Python program to swap two variables?

```python
a =5 
b =14
def swap(p, q):
  temp =p
  p= q
  q = temp 
  return p,q


# Before swapping
print(f'a & b  before : ${a} & ${b}')

a, b = swap(a,b)

# After swapping
print(f'a & b  after: ${a} & ${b}')
```
