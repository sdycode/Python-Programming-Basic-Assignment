1. Write a Python Program to Display Fibonacci Sequence Using Recursion?

```python
def fib(a,b,c):
  a,b =  b,a+b
  if(c>0):
    c -= 1
    print(a)
    fib(a,b,c)
  else:
    return
fib(1,1,20)
```

2. Write a Python Program to Find Factorial of Number Using Recursion?

```python
def fact(n):
  if(n == 0):
    return 1;
  else:
    return n*fact(n-1)

print(fact(4))
print(fact(6))
print(fact(5))
```

5. Write a Python Program for cube sum of first n natural numbers?

```python
def getCubeSum(n):
  sum = 0; 
  for i in range(1,n+1):
    sum += i**3
  return sum

print(getCubeSum(10))
print(getCubeSum(24))
```
