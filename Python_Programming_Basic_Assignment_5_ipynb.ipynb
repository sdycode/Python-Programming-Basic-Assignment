1. Write a Python Program to Find LCM?

```python
# importing the module
import math

# function to calculate LCM
def LCMofList(a):
  lcm = a[0]
  for i in range(1,len(a)):
    lcm = lcm*a[i]//math.gcd(lcm, a[i])
  return lcm
```

2. Write a Python Program to Find HCF?

```python
def find_gcd(x, y):
    while(y):
        x, y = y, x % y
  
    return x
      
      
l = [2, 4, 6, 8, 16]
  
m=l[0]
n=l[1]
gcd=find_gcd(m,n)
  
for i in range(2,len(l)):
    gcd=find_gcd(gcd,l[i])
      
print(gcd)
```

5. Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python


def add(x, y):
    return x + y


def subtract(x, y):
    return x - y

def multiply(x, y):
    return x * y

def divide(x, y):
    return x / y


print("Prform follwoing operations")
print(" + Add")
print(" - Subtract")
print(" * Multiply")
print(" / Divide")

while True:

    opearation = input("Choose opeation by entering ( + - * / )")


    if opearation in ('+', '-', '*', '/'):
        m = float(input("Enter first number: "))
        n = float(input("Enter second number: "))

        if opearation == '+':
            print(m, "+", n, "=", add(m, n))

        elif opearation == '-':
            print(m, "-", n, "=", subtract(m, n))

        elif opearation == '*':
            print(m, "*", n, "=", multiply(m, n))

        elif opearation == '/':
            print(m, "/", n, "=", divide(m, n))
        
   
        do_again = input("Do you try more: Y / N ")
        if do_again == "N":
          break
    
    else:
        print("Incorrect Input")
```
