Write a Python Program to find sum of array?
-> 

```python
l = [5,8,6,3,8,1]

def giveSum(l):
  sum=0
  for i in l:
    sum += i
  return sum

giveSum(l)
```

Write a Python Program for array rotation?

```python
def rightRot(li,n):
  l = list(li)
  for i, e in enumerate(li):
   
    l[int((i+n) % len(l))]= e
  return l

l = [5,8,6,3,8,1]
rightRot(l,2)
```

```python
def leftRot(li,n):
  l = list(li)
  for i, e in enumerate(li):
    j = i-n
    l[len(l)+j if j < 0 else j]= e
  return l

l = [5,8,6,3,8,1]
leftRot(l,2)
```
